## Total Ratings by Category

```python
categorywise_rating=data.groupby('category')['total ratings'].mean()
categorywise_rating
```

```python
plt.figure(figsize=(18,10))
plt.bar(categorywise_rating.index,categorywise_rating.values)
plt.title("Total Ratings by Game Category",size=30)
plt.xlabel("category of games",size=20)
plt.ylabel("Rating",size=20)
plt.xticks(size=20,rotation=80)
plt.yticks(size=20)
plt.show()
```

## Number of Game Installations by Game Category

```python
def in_thousand (inst):
    if inst == '500.0 k':
        return '0.5 M' 
    elif inst == '100.0 k':
        return '0.1 M'
    else:
        return inst
data['installs']= data['installs'].apply(in_thousand)
data['installs']= data['installs'].str.replace( 'M', '').str.strip().astype('float')
data= data.rename(columns={'installs':'installs_in_million'})
data['installs_in_million'].value_counts()
```

```python
install_by_category = data.groupby('category')['installs_in_million'].mean()
install_by_category
```

```python
plt.figure(figsize=(18,10))
plt.bar(install_by_category.index,install_by_category.values,color="orange")
plt.title("Number of Game Installations by Game Category",size=30)
plt.xlabel("category of games",size=20)
plt.ylabel("instals(million)",size=20)
plt.xticks(size=20,rotation=80)
plt.yticks(size=20)
plt.show()
```

## Paid and free games

```python
plt.pie(data['paid'].value_counts().values,labels=["free","paid"],colors=["green","red"],autopct='%1.1f%%')
plt.title('Paid & Free Games')
plt.show()
```

## Growth in 60 days by Game Category

```python
growth_by_category_60 = data.groupby('category')['growth (60 days)'].mean()
growth_by_category_60
```

```python
plt.figure(figsize=(18,10))
plt.bar(growth_by_category_60.index,growth_by_category_60.values,color="hotpink")
plt.title("Growth in 60 days by Game Category",size=30)
plt.xlabel("category of games",size=20)
plt.ylabel("growth",size=20)
plt.xticks(size=20,rotation=80)
plt.yticks(size=20)
plt.show()
```
